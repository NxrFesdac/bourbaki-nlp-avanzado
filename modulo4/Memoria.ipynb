```python
from openai import OpenAI
import getpass

# Configura el cliente de OpenAI (Reemplaza con tu clave API real)
key = getpass.getpass("Ingresa tu clave API de OpenAI: ")
client = OpenAI(api_key=key)

# ---------------------------------------------------------
# 1. Construyendo memoria de corto plazo
# ---------------------------------------------------------
print("💬 --- 1. CHAT CON MEMORIA A CORTO PLAZO ---")

short_term_memory = [
    {"role": "system", "content": "Eres un asesor bancario profesional, amable y persuasivo. Tu objetivo es entender las necesidades financieras del cliente."}
]

# --- Turno 1 ---
user_input_1 = "Hola, soy Carlos. Estoy pensando en cambiar de banco."
print(f"👤 CLIENTE: {user_input_1}")
short_term_memory.append({"role": "user", "content": user_input_1})

response_1 = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=short_term_memory
).choices[0].message.content

print(f"🏦 BANCO (IA): {response_1}")
short_term_memory.append({"role": "assistant", "content": response_1})
```

```python
# --- Turno 2 ---
user_input_2 = "Pues mi meta principal es ahorrar para el enganche de una casa en unos 2 años. También me vendría bien una tarjeta de crédito, pero odio pagar anualidad."
print(f"\n👤 CLIENTE: {user_input_2}")
short_term_memory.append({"role": "user", "content": user_input_2})

response_2 = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=short_term_memory
).choices[0].message.content

print(f"🏦 BANCO (IA): {response_2}")
short_term_memory.append({"role": "assistant", "content": response_2})
```

```python

# ---------------------------------------------------------
# 2. CREANDO LA MEMORIA A LARGO PLAZO (Extrayendo el perfil del cliente)
# El chat termina. Usamos la IA para leer el historial y extraer
# los datos clave para el CRM (base de datos) del banco.
# ---------------------------------------------------------
print("\n🧠 --- 2. EXTRACCIÓN DE DATOS (CONSTRUYENDO MEMORIA A LARGO PLAZO) ---")

chat_log = str(short_term_memory)

extraction_response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "Eres un analista de datos bancarios. Lee este registro de chat y extrae únicamente los datos clave del cliente y sus metas financieras en una sola oración concisa."},
        {"role": "user", "content": chat_log}
    ]
)

long_term_memory = extraction_response.choices[0].message.content
print("💾 Guardado en el CRM del Banco (Memoria a Largo Plazo):", long_term_memory)
```

```python
# ---------------------------------------------------------
# 3. UN NUEVO DÍA: OFRECIENDO SERVICIOS CON CONTEXTO
# Pasan los días. El cliente abre la app del banco. Borramos la
# memoria a corto plazo, ¡pero inyectamos su perfil financiero!
# ---------------------------------------------------------
print("\n🌅 --- 3. NUEVA SESIÓN (USANDO LA MEMORIA A LARGO PLAZO) ---")

new_short_term_memory = [
    # La IA recibe el perfil financiero del cliente en secreto
    {"role": "system", "content": f"Eres un asesor bancario proactivo. Perfil del cliente actual: {long_term_memory}. Usa esta información para ofrecer un producto bancario específico de manera natural."},

    # El usuario simplemente saluda
    {"role": "user", "content": "Hola, ¿tienen alguna oferta para mí hoy?"}
]

print("👤 CLIENTE: Hola, ¿tienen alguna oferta para mí hoy?")

final_response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=new_short_term_memory
).choices[0].message.content

print(f"🏦 BANCO (IA): {final_response}")
```
